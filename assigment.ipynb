```python
#ABMI/00925/2023
#SHARON WAMBUI GICHUHI

class Product:
    def _init_(self, productID, name, price):
        self.productID = productID
        self.name = name
        self.price = price

    def find_price(self):
        return self.price


class DiscountedProduct(Product):
    def _init_(self, productID, name, price, discount):
        self.productID = productID
        self.name = name
        self.price = price
        self.discount = discount

    def find_price(self):
        return self.price - (self.price * self.discount / 100)

products = []

def add_product():
    productID = input("enter product ID: ")
    name = input("enter product name: ")
    price = float(input("Input the product price: "))
    discount = float(input("Discount percentage : "))

    if discount > 0:
        item = DiscountedProduct(productID, name, price, discount)
    else:
        item = Product(productID, name, price)

    products.append(item)
    print("Product has been added")


def view_products():
    if not products:
        print("No products available.")
        return

    for p in products:
        print(f"Product ID: {p.productID}")
        print(f"Name: {p.name}")
        print(f"Final Price:{p.find_price(): }")


def search_product():
    productID = input("Enter product ID to search: ")
    for p in products:
        if p.productID == productID:
            print(f"Name: {p.name}")
            print(f"Final Price:{p.find_price(): }")
            return
    print("the product has not been found ")


def update_product():
    productID = input("Product ID to update: ")
    for p in products:
        if p.productID == productID:
            price = input("Enter the new price : ")
            if price:
                p.price = float(price)
            print("Updated")
            return
    print("the product has not been found ")


def purchase():
    cart = []
    while True:
        view_products()
        productID = input("Enter Product ID to buy (or 'done' to finish): ")
        if productID () == "done":
            break

        for p in products:
            if p.productID == productID:
                cart.append(p)
                print(f"you added {p.name} to cart ")
                break
        else:
            print("the product has not been found ")

    if not cart:
        print("there are no items in the cart.")
        return

    total = sum(p.find_price() for p in cart)
    print(f"\nCart total: {total:.2f}")
    print("Payment methods: 1. Cash 2. Card 3. M-Pesa")

    method = input("Choose method of payment: ")
    pay = float(input("Enter amount: "))

    if pay >= total:
        print(f"Payment received. Change: {pay - total: }")
    else:
        print("payment is not enough ")

def menu():
    while True:
        print("Retail Shop Menu")
        print("1. Add Product")
        print("2. View Products")
        print("3. Search Product")
        print("4. Update Product")
        print("5. Purchase")
        print("6. Exit")

        choice = input("Choose: ")
        if choice == "1":
            add_product()
        elif choice == "2":
            view_products()
        elif choice == "3":
            search_product()
        elif choice == "4":
            update_product()
        elif choice == "5":
            purchase()
        elif choice == "6":
            print("Thank you for shopping with us ")
            break
        else:
            print("Invalid choice ")

menu()
```
